# Lab 3 - PCA on Iris Dataset
## Imports

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from warnings import warn
from tiblib.visualization import plot_2d
from tiblib.preprocessing import PCA
```

## Loading

```python
iris = pd.read_csv('../datasets/iris.csv',header=None)
X = iris.iloc[:,:-1]
y = iris.iloc[:,-1]
```

## PCA

```python
pca = PCA(n_dims=2)
X_reduced = pca.fit_transform(X)
plot_2d(X_reduced,y)
```
